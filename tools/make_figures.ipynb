```python
import os
import cv2

from matplotlib import pyplot as plt

from lared_dataset.tools.visualizer import TrackAnnotator
from lared_dataset.constants import raw_videos_path, processed_pose_path
```

```python
    cam = 2
    video_path = os.path.join(raw_videos_path, f'cam{cam}.mp4')
    cap = cv2.VideoCapture(video_path)
    tracks = os.path.join(processed_pose_path, 'tracks', f'cam{cam}_final.pkl')
    annot = TrackAnnotator(cap, tracks, None)
```

```python
def make_figure():
    annot.load_frame(200000)
    annot.make_next_frame()
    im_rgb = cv2.cvtColor(annot.out_frame, cv2.COLOR_BGR2RGB)
    fig, ax = plt.subplots(1,1)
    ax.imshow(im_rgb)
    return fig, annot.out_frame
```

```python
fig, cv2_image = make_figure()
```

```python
cv2.imwrite('../paper/images/scene_cam2_2.png', cv2_image)
```
